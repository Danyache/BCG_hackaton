```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/invitro_train.csv')
df
```

```python
df['Значение'] = df['Значение'].replace({
    'Is not detected': 'НЕ ОБНАР',
    'Detected': 'ОБНАРУЖ.',
    'ОБНАРУЖ': 'ОБНАРУЖ.',
    'СМ. КОММ': 'СМ.КОММ.'
})
df['Дата взятия'] = pd.to_datetime(df['Дата взятия'])
df['date_day'] = df['Дата взятия'].dt.round('1d')
df['is_pos'] = (df['Значение'] == 'ОБНАРУЖ.').astype(int)
df
```

```python
df = df.groupby(['Регион', 'Имя теста', 'date_day'])['is_pos'].agg(['sum', 'count']).reset_index()
df
```

```python
df.to_csv('data/data_preprocessed.csv', index=False)
```
