```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set()

%matplotlib inline
```

```python
df_sum = pd.read_csv('submissions/df_submission_sum_final.csv', index_col='Unnamed: 0')
df_sum
```

```python
df_sum.max()
```

```python
df_count = pd.read_csv('submissions/df_submission_count_final.csv', index_col='Unnamed: 0')
df_count
```

```python
df_count.max()
```

```python
df_count.drop('region', axis=1, inplace=True)
df = pd.concat([df_sum, df_count], axis=1)
df
```

```python
test_cols = [
    'Anaplasma Phagocytophillum(кач) ДНК',
    'Borrelia burgdorferi s.l (кач) ДНК',
    'Borrelia miyamotoi (кач) ДНК',
    'Ehrlichia muris/chaffeеnsis(кач) ДНК',
    'Вирус клещевого энцефалита(кач) РНК'
]

for i in range(len(df)):
    row = df.iloc[i]
    for col in test_cols:
        col_sum = col + ' sum'
        col_count = col + ' count'
        if row[col_sum] > row[col_count]:
            print(row['region'], col, row[col_sum], row[col_count])
```

```python
df.loc[
    df['Вирус клещевого энцефалита(кач) РНК sum'] > df['Вирус клещевого энцефалита(кач) РНК count'],
    'Вирус клещевого энцефалита(кач) РНК sum'
] = 0.0

for i in range(len(df)):
    row = df.iloc[i]
    for col in test_cols:
        col_sum = col + ' sum'
        col_count = col + ' count'
        if row[col_sum] > row[col_count]:
            print(row['region'], col, row[col_sum], row[col_count])
```

```python
for col in test_cols:
    col_sum = col + ' sum'
    col_count = col + ' count'
    col_pos_perc = col + ' pos perc'
    df[col_pos_perc] = (df[col_sum] / df[col_count]).fillna(0.0) * 100
df
```

```python
df.isnull().sum()
```

```python
df.fillna(0.0, inplace=True)
df.isnull().sum()
```

```python
df.drop([col for col in df.columns if col.endswith('sum')], axis=1, inplace=True)
df
```

```python
df.max()
```

```python
sub_columns = ['region', 'Anaplasma Phagocytophillum(кач) ДНК count',
       'Anaplasma Phagocytophillum(кач) ДНК pos perc',
       'Borrelia burgdorferi s.l (кач) ДНК count',
       'Borrelia burgdorferi s.l (кач) ДНК pos perc',
       'Borrelia miyamotoi (кач) ДНК count',
       'Borrelia miyamotoi (кач) ДНК pos perc',
       'Ehrlichia muris/chaffeеnsis(кач) ДНК count',
       'Ehrlichia muris/chaffeеnsis(кач) ДНК pos perc',
       'Вирус клещевого энцефалита(кач) РНК count',
       'Вирус клещевого энцефалита(кач) РНК pos perc']
df = df[sub_columns]
df.round().to_csv('submissions/tft_submission_final.csv')
```

# Analysis

```python
for region in df['region'].unique():
    df_region = df[df['region'] == region].round()
    plt.figure(figsize=(11, 8))
    plt.title(region)
    for col in test_cols:
        col_count = col + ' count'
        col_pos_perc = col + ' pos perc'
        plt.plot(df_region.index, df_region[col_count], label=col_count)
        plt.plot(df_region.index, df_region[col_pos_perc], label=col_pos_perc)
    plt.legend(loc='best')
    plt.xticks(rotation=90)
    plt.show()
```
